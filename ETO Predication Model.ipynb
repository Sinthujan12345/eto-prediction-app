```python
from sklearn.model_selection import train_test_split
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20, random_state=0)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
def evaluate_model(model, x_train, y_train, x_test, y_test):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    acc = accuracy_score(y_test, y_pred)

    return acc
```

```python
rf  = RandomForestClassifier(n_estimators=100, criterion="entropy", random_state=0)

knn = KNeighborsClassifier(n_neighbors=5)

xgb = XGBClassifier(
    objective='binary:logistic',
    n_estimators=100,
    random_state=0
)

svm = SVC(kernel="rbf", decision_function_shape="ovr", random_state=0)

gnb = GaussianNB()
```

```python
results = []

# Random Forest
acc = evaluate_model(rf, x_train, y_train, x_test, y_test)
results.append(["Random Forest", acc])

# KNN
acc = evaluate_model(knn, x_train_scaled, y_train, x_test_scaled, y_test)
results.append(["KNN", acc])

# XG Boost
xgb = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=4,
    objective="multi:softmax",
    eval_metric="mlogloss",
    random_state=0
)
results.append(["XG Boost", acc])

# SVM
acc = evaluate_model(svm, x_train_scaled, y_train, x_test_scaled, y_test)
results.append(["SVM", acc])

# Gaussian Naive Bayes 
acc = evaluate_model(gnb, x_train, y_train, x_test, y_test)
results.append(["Gaussian NB", acc])
```

```python
results_df = pd.DataFrame(results, columns=["Model", "Accuracy"])
results_df
```
